```python
from matplotlib.axis import YTick

#method 2 to load the data in the form of arrays -by library numpy

import numpy as np

x1=np.array(df[['Gender','EstimatedSalary','Age']])

y=np.array(df[['Purchased']])
```

```python
#importing standard scalar from scikit-learn
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x1= sc.fit_transform(x1)

#x_test = sc.transform(x_test)
```

```python
#Understanding the distribution with seaborn
with sns.plotting_context("notebook",font_scale=2.5):
  g=sns.pairplot(df[['Gender','Age','EstimatedSalary','Purchased']],hue='Purchased',palette='tab20',size=6)
g.set(xticklabels=[]);
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
WIDTH=28
HEIGHT=6

def plot_vs(frame,col1,col2,hue):
  plt.figure(figsize=(WIDTH, HEIGHT*2))
  sns.scatterplot(data=frame,x=col1,y=col2,hue=hue)
  plt.title(f"Scatter plot | {col1} vs {col2}")
  plt.show()
```

```python
plot_vs(df,'Age','EstimatedSalary','Gender')
plot_vs(df,'Age','EstimatedSalary','Gender')
```

```python
# EstimatedSalary : Outliers handling
df = df.drop(df.query("EstimatedSalary > 120000 and Purchased == 0").index)# Age : Outliers handling
df = df.drop(df.query("Age > 55 and Purchased == 0").index)
```

```python
#splitting the data into training and testing set
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x1_train,y_train)
```

```python

y_pred=model.predict(x1_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))#predictated and actual value
```

```python
print("predicted value for training data:",model.score(x1_train,y_train))
```

```python
print("predicted value for testing data:",model.score(x1_test,y_test))
```

```python
accuracy=model.score(x1,y)*100
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)

print(cm)

accuracy_score(y_test, y_pred)
```
